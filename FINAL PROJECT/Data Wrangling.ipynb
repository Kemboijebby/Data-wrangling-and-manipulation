 2.DEALING WITH MISSING VALUES: Most of the datasets having a vast amount of data contain missing values of NaN; They are taken care of by replacing them with the mean, mode and the most frequent value of the column or simply dropping the row having a NaN value.

 3.RESHAPING DATA; In the process, data is manipulated according to the requirements, where new data can be added or pre-existing data can be modified.

 4.FILTERING DATA;Sometimes datasets are copmrised of unwanted rows or columns which are required to be removed or filtered.

 5.After dealing with the row datasets with the above functionalities we get an efficient dataset as per our requirements and then it can be used for a required purpose like data anlyzing,machine learning,data visualization, model training etc.
